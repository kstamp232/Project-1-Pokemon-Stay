## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics of all their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def player_power(PLAYERS, POKEDEX, player_id):
    
# place player pokemon into variable to make it easier to iterate through
    pokemon_in_player = PLAYERS[player_id]['player_pokemon']
    player_pow = 0
    
# loop through each pokemon held by each player and add defense / attack values
# in order to allow for addition I turned the values to integers
# print players name and total of powers from previous addition
    for pokemon in range(len(pokemon_in_player)):
        player_pow = int(pokemon_in_player[player_id]['attack'])+int(pokemon_in_player[player_id]['defense'])+int(pokemon_in_player[player_id]['special_attack'])+int(pokemon_in_player[player_id]['special_defense'])
        print(PLAYERS[player_id]['player_name'], "'s power is", player_pow)
        return player_pow
    
player_power(PLAYERS, POKEDEX, 1)
player_power(PLAYERS, POKEDEX, 2)
````

````python
# I changed data into a more read-able list by replacing the " " and removing the ''\n' using .split

pd_no_quotes = raw_pd.replace('"', '')
pd_split = pd_no_quotes.split('\n')
````

````python
clean = []

# for each line in the data above, split on the ',' to create a new line and append to a new list
for a in range(len(pd_split)):
    pd_split2 = pd_split[a].split(",")
    clean.append(pd_split2)
    
# iterate through the new lines in the new list
# try changing the item to a float (meaning it is a number)
# if not then keep as is (for the letters)
    for l in range(len(clean)) :
        for i in range(len(clean[l])) :
            try :
                clean[l][i] = float(clean[l][i])
            except :
                    clean[l][i] = clean[l][i]
````
